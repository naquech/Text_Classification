# Text Classification / NLP

Sentiment analysis focuses on analyzing and undertsanding emotions from subjective text patterns; it identifies opinions and attitudes towards certain topics and is usefuls in classifying viewpoints as positive or negative. 

The data set for this project is sourced from <a href='https://www.figure-eight.com/data-for-everyone/'> Figure Eight</a> platform and the name of the set is Sentiment Analysis on Global Warming/Climate Change. The set contains three classifications: Yes if the tweet suggests global warming existence, No if the tweet is in disagreement, and neutral if the tweet is not engaged on either side. A column with the confidence score for the classification of each tweet is included in the set with a total of 6090 rows.

# Feature Engineering for Text Data

The whole dataset will be cleaned and prepared, after this process I will make the split into train and test data.

* Lowercase all text
* Remove stop words, punctuation and special characters
* Tokenize data: split a string of text into individual tokens (words)
* Remove suffixes using stemming technique

```python
import nltk
from nltk.corpus import stopwords
from nltk.collocations import *
from nltk import word_tokenize
import string, re
```

Print some tweets to look at what needs to be extracted:

```python
print(data['tweet'][5], '\n')
print(data['tweet'][10], '\n')
print(data['tweet'][25], '\n')
print(data['tweet'][50], '\n')
print(data['tweet'][500], '\n')
print(data['tweet'][1236], '\n')
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt
```

```python
#remove @users

data['tweet'] = np.vectorize(remove_pattern)(data['tweet'], "@[\w]*") 
print(data['tweet'][5], '\n')
```

```python
#remove puntuaction

data['tweet'] = data['tweet'].str.replace("[^a-zA-Z#]", " ")
print(data['tweet'][5], '\n')
```

```python
# lowercase, remove stop and short words

stopwords_list = stopwords.words('english')
stopwords_list += ['link']
```

```python
def process_article(tweet):
    tokens = nltk.word_tokenize(tweet) 
    stopwords_removed = [token.lower() for token in tokens if token not in stopwords_list and len(token) > 4]
    return stopwords_removed
```

```python
processed_data = list(map(process_article, data['tweet']))
```

Overview of tokenized and cleaned data

```python
print(processed_data[5], '\n')
print(processed_data[10], '\n')
print(processed_data[25], '\n')
print(processed_data[50], '\n')
print(processed_data[500], '\n')
print(processed_data[1236], '\n')
```

De-tokenization and plug sentences back to dataframe to create train, test set, and some visualizations.

```python
detok_set = [] 
for i in range(len(data)): 
    t = ' '.join(processed_data[i]) 
    detok_set.append(t) 

data['tweet'] = detok_set
```

### Stemming

Stemming the data will reduce dimensionality of the model and will allow us to train on unique words.

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

def stem_sentences(sentence):
    tokens = sentence.split()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

data['tweet'] = data['tweet'].apply(stem_sentences)
```

# Model Evaluation

```python
from keras import models, optimizers, layers
from keras.models import Model, Input, Sequential
from keras.layers import Dense, Input, LSTM, GRU, Embedding, Dropout, SpatialDropout1D, Activation
from keras.layers import Conv1D, Bidirectional, GlobalMaxPool1D, MaxPooling1D
import matplotlib.pyplot as plt
```

### RNN - LSTM

```python
labels_train = data_train['existence'].values
labels_test = data_test['existence'].values
```

```python
y_train = pd.get_dummies(labels_train).values
y_test = pd.get_dummies(labels_test).values
```

```python
tokenizer = text.Tokenizer(num_words=10000)
tokenizer.fit_on_texts(list(data_train.tweet))
list_tokenized_train = tokenizer.texts_to_sequences(data_train.tweet)
x_train = sequence.pad_sequences(sequences=list_tokenized_train, maxlen=100)
```

```python
tokenizer = text.Tokenizer(num_words=10000)
tokenizer.fit_on_texts(list(data_test.tweet))
list_tokenized_test = tokenizer.texts_to_sequences(data_test.tweet)
x_test = sequence.pad_sequences(sequences=list_tokenized_test, maxlen=100)
```

```python
embedding_size = 128
input_ = Input(shape=(100,))
x = Embedding(input_dim=20000, output_dim=embedding_size)(input_)
x = Bidirectional(LSTM(units=25, return_sequences=True))(x)
x = GlobalMaxPool1D()(x)
x = Dropout(rate=0.5)(x)
x = Dense(units=50, activation='relu')(x)
x = Dropout(rate=0.5)(x)
x = Dense(units=3, activation='sigmoid')(x)

lstm_model = Model(inputs=input_, outputs=x)
```

```python
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
lstm_model.summary()
```

```python
lstm_hist = lstm_model.fit(x=x_train, y=y_train, batch_size=32, epochs=20, validation_split=0.1)
```

```python
# Evaluation train set

lstm_results = lstm_model.evaluate(x_train, y_train)
lstm_results
```

```python
# evaluation test set

lstm_test_results = lstm_model.evaluate(x_test, y_test)
lstm_test_results
```

```python
# Model Predictions

y_pred = lstm_model.predict(x_test)
y_pred
```

```python
lstm_dict = lstm_hist.history
lstm_dict.keys()
```

```python
# loss vs the number of epochs

loss = lstm_dict['loss']
epochs = range(1, len(loss)+1)
plt.plot(epochs, loss)

plt.title('Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.show()
```

```python
#training accuracy vs the number of epochs

acc = lstm_dict['acc']
epochs = range(1, len(acc)+1)
plt.plot(epochs, acc)

plt.title('Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

#### Confusion Matrix Plot Function

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import itertools
```

```python
def show_cf(y_true, y_pred, class_names=None, model_name=None):
    cf = confusion_matrix(y_true, y_pred)
    plt.imshow(cf, cmap=plt.cm.Blues)
    
    if model_name:
        plt.title("Confusion Matrix: {}".format(model_name))
    else:
        plt.title("Confusion Matrix")
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
    
    tick_marks = np.arange(len(class_names))
    if class_names:
        plt.xticks(tick_marks, class_names)
        plt.yticks(tick_marks, class_names)
    
    thresh = cf.max() / 2.
    
    for i, j in itertools.product(range(cf.shape[0]), range(cf.shape[1])):
        plt.text(j, i, cf[i, j], horizontalalignment='center', color='white' if cf[i, j] > thresh else 'black')

    plt.colorbar()
```
